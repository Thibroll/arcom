```python
import pandas as pd
import plotly.express as px
```

```python
arcom_df = pd.read_csv("/Users/thibaultrolland/Projects/arcom/data/cross_arcom_off/arcom_data_with_nutriscore_with_noscore.csv")
```

```python
arcom_df['nutriscore_inferred'].value_counts()
```

```python
noscore_df = arcom_df[[x in ['unknown', 'not-applicable'] for x in arcom_df['nutriscore_inferred']]]
```

```python
category_counts = noscore_df['Variété'].value_counts()
sorted_categories = category_counts.index.tolist()

fig = px.histogram(noscore_df,
            x='Variété',
            title=f'Count of most represented ads for products with no nutriscore',
            color_discrete_sequence=px.colors.sequential.Jet,
            height=800,
            category_orders={'Variété': sorted_categories})
fig.show()
```
