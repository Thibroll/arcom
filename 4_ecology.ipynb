```python
import pandas as pd
import plotly.express as px
from datetime import datetime
from unidecode import unidecode
import numpy as np
```

```python
arcom_df = pd.read_csv("/Users/thibaultrolland/Projects/arcom/data/ARCOM.csv")
```

```python
arcom_df.columns
```

```python
arcom_df['script_decoded'] = arcom_df['Script'].apply(lambda x: unidecode(str(x)).lower())
arcom_df['date_formatted'] = arcom_df['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y'))
arcom_df['6_month_interval'] = arcom_df['date_formatted'].dt.to_period('6M')
```

# Top keywords for ecology from chatgpt

```python
kw_list = [
    "durabilite",
    "ecologique",
    "respectueux de l'environnement",
    "biologique",
    "recyclage",
    "energie renouvelable",
    "zero déchet",
    "eco-responsable",
    "biodiversite",
    "conservation",
    "emissions de carbone",
    "emission carbone",
    "carbone neutre",
    "agriculture biologique",
    "eco-friendly",
    "eco friendly",
    "developpement durable",
    "conservation de l'eau",
    "gestion des dechets",
    "mobilite verte",
    "eco-innovation",
    "impact environnemental",
    "responsable"
]


plot_df = arcom_df[[any([kw in str(x).lower() for kw in kw_list])
                    for x in arcom_df['script_decoded']]]

plot_df['kw_eco'] = plot_df['script_decoded'].apply(lambda script: [kw for kw in kw_list if kw in str(script).lower()])
```

```python
fig = px.histogram(arcom_df, x='date_formatted', color='Secteur')
fig.show()
```

```python
# filtered by keywords
fig = px.histogram(plot_df, x='date_formatted',marginal='rug', color='Secteur')
fig.show()
```

```python
for kw in kw_list:
    fig = px.histogram(plot_df[[kw in kw_eco for kw_eco in plot_df['kw_eco']]], x='date_formatted',marginal='rug', color='Secteur', title=kw)
    fig.show()
```

# By given keyword

```python
arcom_df['script_decoded']
```

```python
def plot_by_kw(kw):
    plot_df = arcom_df[[kw in x for x in arcom_df['script_decoded']]]
    fig = px.histogram(plot_df, x='date_formatted',marginal='rug', color='Secteur', title=kw)
    fig.show()
```

```python
kw = "zero emission"
plot_by_kw(kw)
```
