# BERTopic

```python
import nltk
import re

#download stopwords
#nltk.download('stopwords')

# Set up French stopwords
stop_words = set(nltk.corpus.stopwords.words('french'))

# Function to remove non-letter characters from text
def remove_non_letters(text):
    return re.sub(r'[^a-zA-Z\s]+', '', text)

# Function to preprocess text
def preprocess_text(text):
    # Split text into list of words
    words = text.split()
    # Remove stop words using set difference
    filtered_words = list(set(words).difference(stop_words))
    # Join filtered words back into a string
    filtered_text = ' '.join(filtered_words)
    return filtered_text
```

## Emergence du topic écologie dans le temps

```python
plt.clf()

num_occ_eco = ec_df.groupby(['year']).size().reset_index(name='count')
num_occ_all = df.groupby(['year']).size().reset_index(name='count')

# Plotting
plt.figure(figsize=(8, 6))

# Plotting
plt.bar(num_occ_all['year'].astype(str), num_occ_all['count'], label='Overall ads counts', color='orange')
plt.bar(num_occ_eco['year'].astype(str), num_occ_eco['count'], label='Ads with ECOLOGY as keyword', color='blue')

plt.title('Yearly count of ads')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.clf()

num_occ_eco = ec_df.groupby(['year']).size().reset_index(name='count')
num_occ_all = df.groupby(['year']).size().reset_index(name='count')

plt.figure(figsize=(8, 6))

plt.plot(num_occ_eco['year'], num_occ_eco['count']/num_occ_all['count'])

plt.title('Proportion of ads with the keyword ECOLOGY by year')
plt.xlabel('Year')
plt.ylabel('ratio')
plt.grid(True)

plt.show()
```

```python
num_occ_w_eco = ec_df.groupby(['year', 'week']).size().reset_index(name='eco')
num_occ_w_all = df.groupby(['year', 'week']).size().reset_index(name='all')

season_df = pd.merge(num_occ_w_eco, num_occ_w_all, left_on=['year', 'week'], right_on=['year', 'week'], how='outer')

# Define a function to map week number to season
def get_season(week):
    if week in range(3, 13):  # Spring: weeks 3-12
        return 'Spring'
    elif week in range(13, 23):  # Summer: weeks 13-22
        return 'Summer'
    elif week in range(23, 33):  # Autumn: weeks 23-32
        return 'Autumn'
    else:  # Winter: weeks 33-52 and 1-2
        return 'Winter'

season_df['season'] = num_occ_w_all['week'].apply(get_season)

# Group by year and season and aggregate column1 and column2
grouped = season_df.groupby(['year', 'season']).agg({'eco': 'sum', 'all': 'sum'})

# Calculate the ratio of aggregated column1 over aggregated column2
grouped['ratio'] = grouped['eco'] / grouped['all']

# Resetting the index after aggregation
grouped.reset_index(inplace=True)

# Plotting ratios by season and year
plt.figure(figsize=(10, 6))
for year in grouped['year'].unique():
    year_data = grouped[grouped['year'] == year]
    plt.plot(year_data['season'], year_data['ratio'], marker='o', label=year)

plt.title('ratio of ads that are flagged ecology over all by Season and Year')
plt.xlabel('Season')
plt.ylabel('Ratio')
plt.grid(True)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
